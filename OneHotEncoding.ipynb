```
import numpy as np
import pandas as pd
```

```
df = pd.read_csv("/content/houseprice_onehot.csv")
dummies = pd.get_dummies(df.town)
dummies
```

```
new_df = pd.concat([df, dummies], axis='columns')
new_df.drop(['town', 'west windsor', 'Unnamed: 3'], axis='columns', inplace=True)
new_df
```

```
X = new_df.drop(['price'], axis="columns")
X
```

```
y = new_df.price
y
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```
model.predict([[3400,0,0]])
```
